## 1. Variables

A **variable** is a name that points at a value. You create one with `=`
(read it as "gets"): the name on the left *gets* the value on the right.

````python
ticker = "AAPL"       # Apple's stock symbol (text)
price = 187.25        # a price (a number)
shares = 10           # how many shares we own (a whole number)

price                 # the last line of a cell is shown automatically
````

## 2. `print()` — show a result

The last line of a cell prints automatically, but if you want to show **several**
things, use `print()`.

````python
print(ticker)
print(price)
print(shares)
````

## 3. Data types

Every value has a **type**. The four you'll meet most:

| Type | Means | Example |
| ---- | ----- | ------- |
| `int` | whole number | `10` |
| `float` | number with a decimal point | `187.25` |
| `str` | text ("string") — always in quotes | `"AAPL"` |
| `bool` | true or false | `True` |

You can ask Python for a value's type with `type(...)`.

````python
print(type(ticker))   # str
print(type(price))    # float
print(type(shares))   # int

market_open = True    # a bool
print(type(market_open))
````

### Numbers do maths

`int` and `float` behave like you'd expect. Here we work out what our shares
are worth.

````python
position_value = price * shares
print(position_value)
````

## 4. f-strings — putting values inside text

An **f-string** is a string with an `f` in front. Anything inside `{curly braces}`
gets replaced by its value. It's the tidiest way to build a sentence from data.

````python
print(f"We own {shares} shares of {ticker} at ${price} each.")
print(f"That position is worth ${position_value:.2f}.")   # :.2f = 2 decimal places
````

## 5. Lists — a row of values

A **list** holds several values in order, inside square brackets `[]`.
Perfect for a set of tickers.

````python
tickers = ["AAPL", "MSFT", "TSLA"]
print(tickers)
print(len(tickers))        # how many items? -> 3
print(tickers[0])          # the FIRST item (counting starts at 0!) -> AAPL
print(tickers[-1])         # the LAST item -> TSLA
````

Lists can grow and change. `.append(...)` adds to the end.

````python
tickers.append("NVDA")     # add a new ticker
print(tickers)
print(tickers[1:3])        # a "slice": items 1 and 2 -> ['MSFT', 'TSLA']
````

## 6. Dictionaries — labelled values

A **dictionary** (`dict`) stores values under **keys** (labels), inside curly
braces `{}`. Think of it as a mini price sheet: each ticker maps to its price.

````python
prices = {
    "AAPL": 187.25,
    "MSFT": 430.50,
    "TSLA": 251.10,
}

print(prices["AAPL"])          # look up Apple's price by its key
prices["NVDA"] = 121.40        # add a new entry
print(prices)
````

You can ask a dict for all its keys or all its values.

````python
print(prices.keys())           # the tickers
print(prices.values())         # the prices
````

## 7. `if` / `else` — making decisions

`if` runs a block **only when** a condition is true. `else` runs when it isn't.
Notice the colon `:` and the indentation (4 spaces) — Python uses indentation to
know what's "inside" the `if`.

````python
price = 187.25

if price > 200:
    print("Expensive!")
else:
    print("Under 200.")
````

You can check more cases with `elif` ("else if").

````python
change = -1.5    # today's price change

if change > 0:
    print("📈 Up today")
elif change < 0:
    print("📉 Down today")
else:
    print("➡️ Flat today")
````

## 8. `for` loops — do something to every item

A **`for` loop** repeats a block once for each item in a list (or dict). Great
for going through every ticker.

````python
for t in tickers:
    print(f"Checking {t}...")
````

Looping over a dictionary with `.items()` gives you the key **and** the value
each time round.

````python
for ticker, price in prices.items():
    print(f"{ticker} is trading at ${price}")
````

## 9. Functions — reusable code

A **function** is a named recipe you can run again and again. You **define** it
with `def`, give it **inputs** (parameters) in the brackets, and it hands back a
result with `return`.

````python
def position_value(price, shares):
    return price * shares

# now "call" it with different inputs:
print(position_value(187.25, 10))
print(position_value(430.50, 4))
````

Functions make your code readable: the name says *what it does*, so you don't
have to re-read the maths every time.
