# 02 · pandas DataFrames 🐼

Real market data comes as a **table**: rows and columns, like a spreadsheet. The
Python library for tables is **pandas**, and its table type is called a
**DataFrame**.

Everything the finance dashboard does — showing prices, filtering days, charting
— happens on a pandas DataFrame. So this notebook is worth taking slowly.

## What you'll learn
- `import pandas as pd`
- Building a small DataFrame from a dictionary
- Peeking at it: `.head()`, `.tail()`, `.shape`, `.columns`, `.describe()`
- Selecting **one column** (a Series) vs **several columns**
- Picking rows with `.iloc` (by position) and `.loc` (by label)
- Adding a new **computed column**

## 1. Import pandas

By convention we import pandas and nickname it `pd`. You only do this once per
notebook.

````python
import pandas as pd

print("pandas version:", pd.__version__)
````

## 2. Build a DataFrame from a dictionary

You'll usually **download** data (next notebook), but let's build a tiny one by
hand first so you can see how it's put together. Each dictionary key becomes a
**column**; each list is that column's values going down the rows.

````python
data = pd.DataFrame({
    "Ticker": ["AAPL", "MSFT", "TSLA", "NVDA", "GOOGL"],
    "Open":   [186.0, 428.0, 250.0, 119.0, 173.0],
    "Close":  [187.25, 430.50, 251.10, 121.40, 175.00],
    "Volume": [52_000_000, 21_000_000, 98_000_000, 310_000_000, 18_000_000],
})

data
````

Notice the **numbers down the left edge** (0, 1, 2, ...). That's the **index** —
pandas gave each row a label automatically. (With real stock data the index will
be the **date** of each trading day.)

## 3. Peek at a DataFrame

With big tables you don't want to print everything. These are the tools you'll
reach for constantly:

- `.head()` — the first 5 rows
- `.tail()` — the last 5 rows
- `.shape` — `(rows, columns)`
- `.columns` — the column names
- `.describe()` — quick stats (count, mean, min, max...) for number columns

````python
data.head()
````

````python
data.tail(2)          # you can ask for a specific number of rows
````

````python
print("Shape (rows, cols):", data.shape)
print("Columns:", list(data.columns))
````

````python
data.describe()       # summary stats for the numeric columns
````

## 4. Select a column — a Series

Grab a single column with **square brackets and its name**. One column on its own
is called a **Series** (a single labelled column of values).

````python
data["Close"]
````

````python
closes = data["Close"]
print(type(closes))        # a Series
print(closes.max())        # handy methods: max, min, mean, sum...
print(closes.mean())
````

### One column vs several columns

Ask for **several** columns by passing a **list** of names inside the brackets —
so you'll see **double** brackets `[[ ... ]]`. The result is a smaller DataFrame.

````python
data[["Ticker", "Close"]]
````

## 5. Pick rows: `.iloc` and `.loc`

Two ways to grab **rows**:

- **`.iloc[...]`** — by **position** (integer), like a list. `.iloc[0]` = first row.
- **`.loc[...]`** — by **label** (whatever the index is). Here the labels are
  still numbers, so they look similar — but with dated stock data `.loc` lets you
  ask for a specific date.

````python
data.iloc[0]          # the FIRST row (position 0)
````

````python
data.iloc[-1]         # the LAST row
````

````python
data.loc[2]           # the row whose index LABEL is 2
````

You can grab a single cell by combining a row and a column.

````python
print(data.loc[0, "Ticker"])     # label 0, column "Ticker" -> AAPL
print(data.iloc[0]["Close"])     # position 0, then its Close  -> 187.25
````

## 6. Add a computed column

You can make a **new column** out of existing ones. pandas does the maths for
every row at once — no loop needed. Let's work out each day's change from open to
close.

````python
data["Change"] = data["Close"] - data["Open"]
data
````

That new `Change` column is the kind of thing your dashboard will compute — for
example to find the days a stock went **up** (Change > 0). More on that in
notebook 04.
